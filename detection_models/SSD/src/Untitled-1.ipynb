```python
import numpy as np
import os
```

```python
base_path = "/media/udayanga/data_2/latest_dataset_2/latest_dataset/"
text_file = "/media/udayanga/data_2/latest_dataset_2/latest_dataset/night_outdoor_and_daytime_train_files_event_based.txt"
with_anns_list = []
count = 0
with open(text_file,"r") as f:
    files = f.readlines()
    files = [file.rstrip() for file in files]
    for file in files:
        data = np.load(os.path.join(base_path,file))
        anns = data["ann_array"]
        if anns.any():
            with_anns_list.append(file)
        else:
            print("no anns for ",file)
        count += 1
print("There are {} anns originally ".format(count))
```

```python
len(with_anns_list)
```

```python
with_anns_list
```

```python
with open(text_file,"w") as f:
    for file in with_anns_list:
        f.writelines(file + "\n")
```

```python
without_lab_list = []
count = 0
with open(text_file,"r") as f:
    files = f.readlines()
    for file in files:
        #print(file)
        count += 1
        if file.startswith("at_"):
            continue
        without_lab_list.append(file)
print("number of anns ",count)
```

```python
without_lab_list
```

```python
text_file = "/media/udayanga/data_2/latest_dataset_2/latest_dataset/night_outdoor_and_daytime_train_files_event_based_nolab.txt"
```

```python
with open(text_file,"w") as f:
    for file in without_lab_list:
        f.writelines(file)
```
